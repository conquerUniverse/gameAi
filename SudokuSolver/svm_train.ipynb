```
os.listdir()

image_path = []
for files in os.listdir(path):
  label = files
  path1 = os.path.join(path,files)
  if os.path.isdir(path1):
    for image in os.listdir(path1):
      if '.jpg' in image:
          image_path.append((os.path.join(path1,image),label))
image_path
```

```
import random

random.shuffle(image_path)
```

```
image_path
```

```
import cv2

X= []
Y= []
for img in image_path:
  img_resized = cv2.imread(img[0])
  img_resized = cv2.resize(img_resized, (32,32))
  gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
  # img_resized = rgb2gray(img_resized)
  X.append(gray.flatten()) 
  Y.append(int(img[1]))

```

```
Y[0]
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.75, random_state=101)
print(len(x_train))
print(len(x_test))
# print(y_train.shape)
# print(y_test.shape)
```

```
from sklearn import svm
from sklearn import metrics

# an initial SVM model with linear kernel   
svm_linear = svm.SVC(kernel='linear')

# fit
svm_linear.fit(x_train, y_train)
```

```
predictions = svm_linear.predict(x_test)
predictions[:10]
```

```
metrics.accuracy_score(y_true=y_test, y_pred=predictions)
```

```
class_wise = metrics.classification_report(y_true=y_test, y_pred=predictions)
print(class_wise)
```

```
svm_rbf = svm.SVC(kernel='rbf')
svm_rbf.fit(x_train, y_train)
```

```
predictions = svm_rbf.predict(x_test)
predictions[:10]
```

```
metrics.accuracy_score(y_true=y_test, y_pred=predictions)
```

```
class_wise = metrics.classification_report(y_true=y_test, y_pred=predictions)
print(class_wise)
```

```
import pickle
filename = 'svm_model3.sav'
pickle.dump(svm_rbf, open(filename, 'wb'))
```
